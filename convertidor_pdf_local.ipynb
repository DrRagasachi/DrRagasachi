```python
# Instala la biblioteca tabula-py
#!pip install tabula-py

# Importa las bibliotecas necesarias
import pandas as pd
import tabula
from google.colab import files

def pdf_a_dataframe(ruta_archivo):
    """
    Extrae tablas de un archivo PDF y las combina en un único DataFrame.

    Parámetros:
        ruta_archivo (str): La ruta del archivo PDF en el sistema.

    Retorna:
        pd.DataFrame: Un DataFrame con las columnas deseadas.
    """
    # Extrae tablas del archivo PDF
    tablas = tabula.read_pdf(ruta_archivo, pages='all')

    # Inicializa un DataFrame vacío para almacenar las tablas combinadas
    dataframe_combinado = pd.DataFrame()

    # Combina todas las tablas en un único DataFrame
    for tabla in tablas:
        df = tabla.copy()
        dataframe_combinado = pd.concat([dataframe_combinado, df], ignore_index=True)

    # Define las columnas deseadas
    columnas_deseadas = ["Fecha", "Descripción", "Origen", "Crédito", "Débito", "Saldo"]

    # Verifica si el DataFrame está vacío
    if dataframe_combinado.empty:
        print("El DataFrame está vacío. No se encontraron datos en el PDF.")
        return pd.DataFrame(columns=columnas_deseadas)  # Retorna un DataFrame vacío con las columnas deseadas

    # Verifica si las columnas deseadas están presentes en el DataFrame
    columnas_disponibles = dataframe_combinado.columns.tolist()
    columnas_finales = [col for col in columnas_deseadas if col in columnas_disponibles]

    # Si no hay columnas finales, retorna un DataFrame vacío con las columnas deseadas
    if not columnas_finales:
        print("No se encontraron las columnas deseadas en el PDF.")
        return pd.DataFrame(columns=columnas_deseadas)

    # Selecciona solo las columnas deseadas
    dataframe_final = dataframe_combinado[columnas_finales]

    # Renombra las columnas para asegurarte de que coincidan con los nombres deseados
    dataframe_final.columns = columnas_deseadas[:len(columnas_finales)]

    return dataframe_final

# Ruta del archivo PDF en tu computador o entorno de Colab
ruta_del_pdf = '/content/BANCO GALICIA 012024.pdf'

# Llama a la función para convertir el PDF en un DataFrame único
dataframe_combinado = pdf_a_dataframe(ruta_del_pdf)

# Verifica si se extrajeron datos
if not dataframe_combinado.empty:
    print("Datos extraídos correctamente:")
    print(dataframe_combinado.head())  # Muestra las primeras filas del DataFrame

    # Guarda el DataFrame en un archivo Excel
    nombre_archivo = "resultados.xlsx"
    dataframe_combinado.to_excel(nombre_archivo, index=False)
    print(f"El archivo '{nombre_archivo}' se ha guardado correctamente.")

    # Descarga el archivo Excel
    files.download(nombre_archivo)
else:
    print("No se pudo extraer ningún dato del PDF.")
```
